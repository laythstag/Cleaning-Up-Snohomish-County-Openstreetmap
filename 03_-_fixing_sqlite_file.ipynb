```python
def format_and_spelling(value, correct, place, to_check, common_mistakes):
    if value in correct[place]: #when value matches a value in to_check, it will return the value
        return value
    elif ',' in value: # if the value has a comma in it, it will split it and check it again.
        split_val = value.split(',')
        if len(split_val) == 2:
            if value not in common_mistakes[place].keys():
                common_mistakes[place][value] = format_and_spelling(split_val[0], correct, place, to_check, common_mistakes)
        else:
            if value not in common_mistakes[place].keys():
                if value not in to_check[place].keys():
                    to_check[place][value] = ''
            elif value in common_mistakes[place].keys():
                return format_and_spelling(common_mistakes[place][value], 
                                           correct, place, to_check, common_mistakes)
    else: # check if the value is in the common_mistakes list and return the corrected value.
        if value not in common_mistakes[place].keys():
            if value not in to_check[place].keys():
                to_check[place][value] = ''
        elif value in common_mistakes[place].keys():
            return format_and_spelling(common_mistakes[place][value], 
                                       correct, place, to_check, common_mistakes)

conn.commit()
```

```python
# Standardizes all postal codes.
def postalcodes(c, table):
    c.execute("""
    SELECT *
    FROM {table}
    WHERE key='postal_code';
    """.format(table=table))
    for f in c.fetchall():
        if '-' in f[2]:
            osmid = f[0]
            key = f[1]
            new_value = f[2].split('-')[0]
            old_value = f[2]
            osmtype = f[3]
            c.execute("""
            UPDATE {table}
            SET id='{osmid}', key='{key}', value='{new_value}', type='{osmtype}'
            WHERE id='{osmid}' and key='{key}' and value='{old_value}' and type='{osmtype}';
            """.format(table=table, osmid=osmid, key=key, new_value=new_value, osmtype=osmtype, old_value=old_value))

postalcodes(c, 'new_nodes_tags')
postalcodes(c, 'new_ways_tags')
conn.commit()
```
